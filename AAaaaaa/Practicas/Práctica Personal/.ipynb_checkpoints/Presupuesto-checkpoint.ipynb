1. Exportar la base de datos que contiene todo mi inventario
2. Limpiar las entradas para solo tener las armas que se puedan subir de nivel y que no hayan sido mejoradas a categoría obra maestra
3. Calcular la cantidad de recursos necesaria para mejorar cada una al nivel deseado y los recursos necesarios para modificar sus perks a ese nivel

```python
# Carga de based e datos
import pandas as pd
import numpy as np
vault = pd.read_csv('destinyWeapons.csv')
```

```python
# Fitlrado de armas crateadas
vault = vault[vault['Crafted'] == True]
print(vault)
# Limpieza de columnas irrelevantes
vault = vault.drop(['Hash', 'Id', 'Tag', 'Source', 'Element', 'Power', 'Year','Season','Power Limit','Equipped','Event',
                    'Guard Endurance','Swing Speed', 'Shield Duration', 'Airborne Effectiveness', 'Foundry', 'Loadouts', 'Notes', 'Owner', 'Locked',
                    'Handling',  'Charge Time', 'Draw Time',  'Accuracy',  'Charge Rate', 'Guard Resistance',  'Guard Efficiency','Zoom','Blast Radius','Velocity','Stability','ROF','Reload','Mag', 'Recoil','AA','Impact','Range',
                    'Perks 0', 'Perks 1', 'Perks 2', 'Perks 3', 'Perks 4', 'Perks 5', 'Perks 6', 'Perks 7', 'Perks 8', 'Perks 9', 'Perks 10', 'Perks 11', 'Perks 12', 'Perks 13', 'Perks 14'], axis=1)
print(vault)
```

```python
# Filtrado de armas sin mejorar
vault = vault[vault['Crafted Level'] < 17]
# Falta comprobar cuales YA HAN sido mejoradas

# Valores que no componen ningun CSV (Manuales, sigh):
glimmer = 233480
legendaryFragments = 44013
enCores = 939+999
enPrism = 28
ascendantAy = 9

totalTeoricalGlimmer = glimmer + ((legendaryFragments/10)*10000)
print("Total de glimmer", totalTeoricalGlimmer)


lowLevels=0
midLevels=0
for i, fila in vault.iterrows():
    level = fila['Crafted Level']
    if level <=10:
        lowLevels+=level
    else:
        lowLevels+=10
        midLevels+=level-10
        
# Cosos que faltan
# Valor de las mejoras de nivel 1-10 y 10-20
print ("Niveles 1-10", lowLevels, ", Niveles 10-16", midLevels)

# Coste de cambiar masterwork, perks de stats y perks de uso

print(vault)
```
